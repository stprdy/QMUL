```python
x = "Hello " + "World"
print(x)
```

## Special characters and escape sequences

```python
print("Single quote \'")
print("Double quote \"")
print("Backslash \\")
print("Bell \a bel?")
print("Backspace\b character")
print("Formfeed \f What's this?")
print("Newline \n I know this!")
print("Carriage-return \r test!")
print("Tab\tthis is a tab\tsee?")
print("Vertical tab? \vIs this spaced?")
```

### Removing extra whitespace

```python
x = "  Hello,     World\t\t"
print(x)
print(x.strip())
print(x.lstrip())
print(x.rstrip())
```

### String searching

```python
x = "Mississippi"
x.find("ss") # returns index of first match
```

```python
x.find("zz")
```

```python
x.count("ss")
```

### Modifying strings

```python
x = "Mississippi"
x.replace("ss", "+++")
```

```python
x = "hello, World!"
print(x)
print(x.lower())
print(x.upper())
print(x.capitalize())
print(x.title())
print(x.swapcase())
```

###  Modifying strings with list manipulations

```python
text = "Hello, World"
print(text)
wordList = list(text)
print(wordList)
wordList[6:] = []
wordList.reverse()
print(wordList)
text = "".join(wordList)
print(text)
```

### Useful methods and constants

```python
x = "123"
print(x)
print(x.isdigit())
print(x.isalpha())
x = "M"
print(x)
print(x.islower())
print(x.isupper())
```

```python
x = ['"abc"', 'def', '"ghi"', '"klm"', 'nop']
x
```

```python
x_new = []
for w in x:
    x_new.append(w.strip('"'))
x_new
```

### Using the format method

```python
"{0} is the {1} of {2}".format("Ambrosia", "food", "the gods")
```

```python
"{food} is the food of {user}".format(food="Ambrosia", user="the gods")
```

### Formatting strings with %

```python
"%s is the %s of %s" % ("Brussels Sprouts", "food", "foolish")
```

```python
x = [1, 2, "three"]
"the %s contains: %s" % ("list", x)
```

```python
print("Pi is <%-6.2f>" % 3.14159)
print("Pi is <%6.2f>" % 3.14159)
print("Pi is <%s>" % 3.14159)
```
